```python
import pandas as pd
import matplotlib.pyplot as plt
from apyori import apriori
from mlxtend.frequent_patterns import fpgrowth
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder
df = pd.read_csv('https://github.com/FlavioAAMotta/didatico/raw/main/CD/2024/Associacao/compras.csv')
df.head()
```

```python
df = df.drop(columns=['cliente'])
```

```python

df.stack().unique()
```

```python
df.isnull().sum()
```

```python
itens_mais_comprados = df.stack().value_counts()
itens_mais_comprados.head()
```

```python
transactions = []
for i in range(df.shape[0]):
    transactions.append([str(df.values[i, j]) for j in range(df.shape[1])])

te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
one_hot_df = pd.DataFrame(te_ary, columns=te.columns_)

# Visualizar os dados one-hot encoded
print(one_hot_df.head())
```

```python
transactions = []
for i in range(0,df.shape[0]):
    transactions.append([str(df.values[i,j])for j in range(0,df.shape[1])])
regras = apriori(transactions=transactions, min_support=0.009, min_confidence=0.2, min_lift=3, min_length=2, max_length=2)
resultados = list(regras)
resultados
```

```python
df_compras_lista = df.apply(lambda x: x.dropna().tolist(), axis=1).tolist()

te = TransactionEncoder()
te_ary = te.fit(df_compras_lista).transform(df_compras_lista)


df_onehot = pd.DataFrame(te_ary, columns=te.columns_)

frequent_itemsets = fpgrowth(df_onehot, min_support=0.009,use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=3)
rules.size
```

```python
rules = rules.sort_values(by='lift', ascending=False)
rules.head(50)
```

```python
rules.tail(50)
```

```python
# Ordenando as regras pelo valor do lift em ordem decrescente
rules_lift_sorted = rules.sort_values(by='lift', ascending=False)

# Selecionando os melhores antecedentes e consequentes baseados no lift
melhores_lift = rules_lift_sorted[['antecedents', 'consequents', 'lift','support' , 'confidence']].head(3)


print("Melhores antecedentes e consequentes baseados no Lift:")
melhores_lift
```

```python
# Ordenando as regras pelo valor da confiança em ordem decrescente
rules_confidence_sorted = rules.sort_values(by='confidence', ascending=False)

# Selecionando os melhores antecedentes e consequentes baseados na confiança
melhores_confidence = rules_confidence_sorted[['antecedents', 'consequents', 'confidence', 'support', 'lift']].head(3)

print("\nMelhores antecedentes e consequentes baseados na Confiança:")
melhores_confidence
```

```python


rules_support_sorted = rules.sort_values(by='support', ascending=False)

melhores_support = rules_support_sorted[['antecedents', 'consequents', 'support','lift','confidence']].head(3)

print("\nMelhores antecedentes e consequentes baseados no Suporte:")
melhores_support
```
