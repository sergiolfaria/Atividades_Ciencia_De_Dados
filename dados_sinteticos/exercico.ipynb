```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('./dados_sinteticos_com_erros_atualizado.csv')
pd.set_option('display.max_columns', None)
df.head()
```

```python
df.isnull().sum()
```

```python
print(df[df.isnull().any(axis=1)])
df = df.dropna()
```

```python
df.isnull().sum()
df.head()
```

```python
df = df.drop(columns=['valor_y','valor','valor_x','tamanho'])
df['Valor real'] = df['total'] / df['quantidade']
df.head(10)
```

```python

df = df.drop_duplicates(subset=['hora', 'cliente', 'produto', 'quantidade','vendedor'])
df.head(100)
```

```python
colunas_selecionadas = ['categoria_x', 'categoria_y', 'categoria']

print(df[colunas_selecionadas].head(50))
print(df[colunas_selecionadas].tail(50))
```

removendo colunas categoria_x  categoria_y já que tudo indica que são completamente iguais

```python
df = df.drop(columns=['categoria_x' ,'categoria_y','pais'])
df.head(20)
```

identificando e removendo alguns outliers

```python
plt.boxplot(df['total'])
plt.title('Boxplot do total ')
plt.xlabel('Variável')
plt.show()
```

```python
#removendo outliers
limite = 4000
df  = df[df['total'] < limite]
plt.boxplot(df['total'])
plt.title('Boxplot do total ')
plt.xlabel('Variável')
plt.show()
```

```python
plt.boxplot(df['Valor real'])
plt.title('Boxplot do Valor real ')
plt.xlabel('Variável')
plt.show()
```

```python
limite = 150
df = valores_filtrados = df[df['Valor real'] < limite]
plt.boxplot(df['Valor real'])
plt.title('Boxplot do Valor real ')
plt.xlabel('Variável')
plt.show()
```
